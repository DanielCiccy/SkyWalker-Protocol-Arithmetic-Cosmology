# Annex: When Negotiation Fails (Systemic Bankruptcies)

The universe is teeming with rogue planets wandering in the void. These are the "sacrifices" of the equation. To understand how a system fails to find its equilibrium, astrophysicists use the **Mutual Hill Radius**.

### 1. The Sphere of Influence (The Hill Radius)
Each planet has a gravitational "bubble" around it, called the Hill sphere. If two planets form too close to each other, their bubbles touch. There is no room left to "negotiate" (resonance becomes impossible).

The mathematical formula for the mutual Hill radius between two planets is:
$$R_{H, mut} = \left( \frac{m_1 + m_2}{3 M_*} \right)^{\frac{1}{3}} \left( \frac{a_1 + a_2}{2} \right)$$
*(Where $m$ represents planetary masses, $M_*$ the stellar mass, and $a$ their distances from the star).*

### 2. The Gladman Criterion (The Boundary of Chaos)
In 1993, physicist Brett Gladman proved mathematically that for two planets to survive without ejecting each other, the distance between their orbits ($\Delta a$) must be strictly greater than **3.46 times** their mutual Hill Radius.
* If $\Delta a > 3.46 R_{H,mut}$ : Equilibrium is found.
* If $\Delta a < 3.46 R_{H,mut}$ : Ballistic chaos ensues. The more massive planet acts as a slingshot, ejecting the other from the stellar system.
