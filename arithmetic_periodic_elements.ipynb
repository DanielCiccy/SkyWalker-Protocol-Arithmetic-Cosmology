##I. The Architecture of Numerical Matter (The Chemical Model)

In this universe, multiplication is no longer a quantitative operation, but a spatial chemical reaction.

* Primordial Nuclei (Rank 0): Pure prime numbers. They are the fundamental building blocks, the protons of the system.

* Noble Gases (Stable Matter): Perfect squares ($p^2$, $p^2q^2$, etc.). These are states of absolute stability, geometrically saturated structures (valence 0) that no longer seek to bond.

* Families (The Periodic Table): Every natural number has a "square-free core". This core defines its valence. All numbers sharing the same valence form a single column in the Mendeleev table. They are "isotopes" seeking to fuse with elements of their own family to reach the state of grace: the perfect square.

```python
# @title
import sympy
from collections import defaultdict
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

def get_square_free_core(n):
    """Extracts the 'square-free core' (the valence/family of the number)."""
    if n == 1:
        return 1
    factors = sympy.factorint(n)
    core = 1
    for prime, exponent in factors.items():
        if exponent % 2 != 0:
            core *= prime
    return core

# ---------------------------------------------------------
# PHASE 1: LARGE SCALE GENERATION
# ---------------------------------------------------------
limit = 10000
families = defaultdict(list)

print(f"Analyzing the first {limit} integers...")

for element in range(1, limit + 1):
    valence = get_square_free_core(element)
    families[valence].append(element)

# ---------------------------------------------------------
# PHASE 2: STATISTICAL ANALYSIS
# ---------------------------------------------------------
# Count the number of elements in each family
family_sizes = {valence: len(members) for valence, members in families.items()}

# Transform into a DataFrame for easy sorting
df_families = pd.DataFrame(list(family_sizes.items()), columns=['Valence', 'Population'])
df_families = df_families.sort_values(by='Population', ascending=False).reset_index(drop=True)

# Key statistics
noble_gases = family_sizes.get(1, 0)
max_family = df_families.iloc[0]
max_valence_found = max(families.keys())

print("-" * 50)
print(f"EXPERIMENT RESULTS (N = {limit})")
print("-" * 50)
print(f"Total distinct families created: {len(families)}")
print(f"Noble Gases Population (Perfect Squares): {noble_gases} elements")
print(f"Most populated family: Family of {max_family['Valence']} (with {max_family['Population']} elements)")
print(f"Heaviest element (core) detected: {max_valence_found}")
print("-" * 50)

# ---------------------------------------------------------
# PHASE 3: SPECTROGRAM (Visualization)
# ---------------------------------------------------------
top_20 = df_families.head(20)

plt.figure(figsize=(12, 6))
bars = plt.bar(top_20['Valence'].astype(str), top_20['Population'], color='teal', edgecolor='black')

# Highlight Noble Gases if they are in the Top 20
for i, bar in enumerate(bars):
    if top_20.iloc[i]['Valence'] == 1:
        bar.set_color('gold')
        bar.set_edgecolor('black')

plt.title(f"Spectrogram of the Top 20 Arithmetic Families (N={limit})", fontsize=14)
plt.xlabel("Family Valence (The 'Core')", fontsize=12)
plt.ylabel("Number of elements in the family", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add custom legend
gold_patch = mpatches.Patch(color='gold', label='Noble Gases (Valence 1)')
teal_patch = mpatches.Patch(color='teal', label='Standard Families')
plt.legend(handles=[gold_patch, teal_patch])

plt.show()
```

```python
# @title
import networkx as nx
import matplotlib.pyplot as plt
import sympy
import math
import matplotlib.cm as cm

def get_square_free_core(n):
    """Returns the square-free core (the Family) of a number."""
    if n == 1: return 1
    core = 1
    for p, e in sympy.factorint(n).items():
        if e % 2 != 0:
            core *= p
    return core

# ---------------------------------------------------------
# 1. CREATION OF THE UNIVERSE AND BONDS
# ---------------------------------------------------------
number_limit = 60  # Going up to 60 for a nice spatial map
G = nx.Graph()

# Add nodes (atoms) and their "Family" attribute
for n in range(1, number_limit + 1):
    family = get_square_free_core(n)
    G.add_node(n, family=family)

# Add edges (stable chemical bonds)
# We link A and B ONLY if A * B = a perfect square
for i in range(1, number_limit + 1):
    for j in range(i + 1, number_limit + 1):
        product = i * j
        root = math.isqrt(product)
        if root * root == product:  # If it's a perfect square
            G.add_edge(i, j)

# ---------------------------------------------------------
# 2. MAPPING AND VISUALIZATION (The Constellation Map)
# ---------------------------------------------------------
plt.figure(figsize=(14, 10))

# Extract families to manage colors
families = [node[1]['family'] for node in G.nodes(data=True)]
unique_values = list(set(families))
color_map = cm.get_cmap('tab20', len(unique_values))

# Assign a specific color to each node based on its family
node_colors = [color_map(unique_values.index(f)) for f in families]

# Force Noble Gases (Family 1) to Yellow/Gold to spot them easily
for idx, f in enumerate(families):
    if f == 1:
        node_colors[idx] = 'gold'

# Calculate spatial layout (linked elements attract each other)
pos = nx.spring_layout(G, seed=42, k=0.5)

# Draw the graph
nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=600, edgecolors='black')
nx.draw_networkx_edges(G, pos, width=2, alpha=0.5, edge_color='gray')
nx.draw_networkx_labels(G, pos, font_size=10, font_family='sans-serif', font_weight='bold')

plt.title("The Collider: Mapping Stable Matter (Covalent Bonds)", fontsize=16)
plt.axis('off') # Remove the frame for a "space" effect
plt.tight_layout()
plt.show()
```

```python
# @title
import numpy as np
import matplotlib.pyplot as plt

# ---------------------------------------------------------
# SIMULATION PARAMETERS
# ---------------------------------------------------------
# Observing evolution over 1000 "time units" or "space units"
N_steps = 1000
t = np.linspace(2, N_steps, N_steps) # Starting at 2 to avoid division by zero

# ---------------------------------------------------------
# MODEL 1: SPACE EXPANSION (The gap between objects)
# ---------------------------------------------------------
# 1. Physical Universe: Accelerated expansion by Dark Energy
# Distance d(t) grows exponentially (Accelerated Hubble's Law)
H = 0.414 # 1e the opposite of the golden number  ($1 + \sqrt{2}$), so.
physical_gap = np.exp(H * t)

# 2. Arithmetic Universe: Gap between two consecutive Noble Gases (perfect squares)
# The distance between n^2 and (n+1)^2 is exactly 2n + 1.
# This is a purely linear growth.
arithmetic_gap = 2 * t + 1

# Normalizing the curves (base 100) so they start from the same origin point
physical_gap_norm = physical_gap / physical_gap[0] * 100
arithmetic_gap_norm = arithmetic_gap / arithmetic_gap[0] * 100

# ---------------------------------------------------------
# MODEL 2: MATTER DENSITY (Dilution of atoms)
# ---------------------------------------------------------
# 1. Physical Universe: Matter dilutes in a (3D) volume growing exponentially
# Density ~ 1 / Volume
physical_density = 1 / (physical_gap**3)
physical_density_norm = physical_density / physical_density[0] * 100

# 2. Arithmetic Universe: Density of fundamental atoms (Prime numbers)
# Obeys the Prime Number Theorem: Density ~ 1 / ln(N)
arithmetic_density = 1 / np.log(t)
arithmetic_density_norm = arithmetic_density / arithmetic_density[0] * 100

# ---------------------------------------------------------
# DASHBOARD VISUALIZATION (Plot)
# ---------------------------------------------------------
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# --- GRAPH 1: The Creation of the Void ---
ax1.plot(t, physical_gap_norm, color='darkred', lw=2, label="Physical Universe (Exponential Acceleration)")
ax1.plot(t, arithmetic_gap_norm, color='darkblue', lw=2, linestyle='--', label="Arithmetic Universe (Linear Growth)")
ax1.set_title("Expansion Speed: The gap between stable elements", fontsize=14)
ax1.set_xlabel("Time (t) / Element index (n)", fontsize=12)
ax1.set_ylabel("Size of the Void (Normalized)", fontsize=12)
ax1.legend()
ax1.grid(True, linestyle=':', alpha=0.7)

# --- GRAPH 2: The Dilution of Matter ---
ax2.plot(t, physical_density_norm, color='darkred', lw=2, label="Physical Universe (Rapid Dilution)")
ax2.plot(t, arithmetic_density_norm, color='darkblue', lw=2, linestyle='--', label="Arithmetic Universe (1/ln(N) - Long survival)")
ax2.set_title("Matter Density: Rarefaction of primordial elements", fontsize=14)
ax2.set_xlabel("Time (t) / Size of the Universe (N)", fontsize=12)
ax2.set_ylabel("Matter density (%)", fontsize=12)
ax2.legend()
ax2.grid(True, linestyle=':', alpha=0.7)

plt.suptitle("Comparative Evaluation: Physical Cosmology vs Arithmetic Cosmology", fontsize=16, fontweight='bold', y=1.05)
plt.tight_layout()
plt.show()
```

 The "Asymmetry Constrictor" (Python Script)This script quantifies the geometric cost of transgression for various atomic bonds.

```python
# @title
import numpy as np
import matplotlib.pyplot as plt

def analyse_transgression(p, q):
    # Angle dyadique idéal (Base 2)
    ideal = np.pi / 4
    # Angle réel de la transgression (pq)
    reel = np.arctan(p / q)
    # Mesure du différentiel par le Cercle Lecteur
    delta_theta = np.degrees(abs(ideal - reel))
    return delta_theta

# Visualisation de la résolution de l'abîme
isotopes = [(3, 5), (5, 7), (11, 13), (17, 19), (71, 73), (197, 199)]
labels = [f"{p}x{q}" for p, q in isotopes]
valeurs = [analyse_transgression(p, q) for p, q in isotopes]

plt.figure(figsize=(10, 5))
plt.plot(labels, valeurs, 'o--', color='#1f77b4', label='Differential asymetry ($\Delta\\theta$)')
plt.fill_between(labels, valeurs, alpha=0.1, color='#1f77b4')
plt.title("Resolution of the Asymmetry Abyss by the Reader Circle")
plt.ylabel("Phase Shift (Degrees)")
plt.xlabel("Transgression Thresholds (pq)")
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

```python
# @title
import numpy as np

def asymmetric_constrictor(p, q):
    """
    The Python 'constrictor' that unfolds the Reader Circle Axiom.
    It measures the exact tension (differential asymmetry) between
    the perfection of base 2 and the anomaly of the semiprime pq.
    """
    # 1. The pure dyadic space: the angle of perfection (Square Diagonal)
    dyadic_angle_rad = np.pi / 4  # Exactly 45°

    # 2. The asymmetric perturbation: the angle of reality (Rectangle p x q)
    pq_angle_rad = np.arctan(p / q)

    # 3. The Circle reads the differential (The abyss of asymmetry Δθ)
    asymmetry_rad = abs(dyadic_angle_rad - pq_angle_rad)
    asymmetry_deg = np.degrees(asymmetry_rad)

    # 4. Display the Circle's reading
    print(f"► READING OF THE ELEMENT [p={p} | q={q} | Isotope={p*q}]")
    print(f"  ├─ Dyadic Norm (Base 2): 45.00°")
    print(f"  ├─ Real Angle of Matter: {np.degrees(pq_angle_rad):.2f}°")
    print(f"  └─ ASYMMETRY READ BY THE CIRCLE (Δθ): {asymmetry_deg:.2f}°\n")

# Let's run the constrictor on several "anomalies" (arithmetic atoms)
# Observe how asymmetry decreases as prime numbers get closer.
pairs_to_read = [
    (3, 5),     # Large gap
    (5, 7),     # Closer
    (11, 13),   # Densification
    (71, 73),   # High stability
    (197, 199)  # Almost a perfect square
]

print("=== THE CIRCLE: DIFFERENTIAL ASYMMETRY READER ===\n")

for p, q in pairs_to_read:
    asymmetric_constrictor(p, q)

print("CONSTRICTOR'S CONCLUSION:")
print("As matter 'ascends' in complexity (large twin primes),")
print("the circle resolves asymmetry (Δθ tends toward 0). The universe tends toward dyadic order.")
```
