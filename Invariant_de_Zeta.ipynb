**Étape 1: Le biais de notre finitude (Le Carré et l'Orthogonalité)**
L'être humain a un besoin vital de limites pour exister, mesurer et agir. Nos outils mentaux sont construits sur la rigidité, la symétrie parfaite et la mesure finie. * Si l'on pousse la puissance géométrique à l'infini ($p \to \infty$), la forme se fige et devient un **carré** parfait, dur et orthogonal.
* Si on l'adoucit pour revenir à l'équilibre matériel ($p = 2$), la forme se lisse et devient un **cercle**.

Le cercle et le carré ne sont donc pas des entités séparées: ils sont les deux extrémités d'une même métamorphose entre l'Infini et le Fini. # 2. La Rupture : L'Ellipse et le vertige de Ramanujan

Nous quittons ici la pureté du cercle pour affronter la complexité de l'ellipse. Dès que la forme s'étire et perd sa symétrie parfaite, la géométrie simple s'effondre.

Pour calculer le contour exact d'une ellipse, la constante $\pi$ ne suffit plus. La machine doit recourir à une **intégrale elliptique** : un calcul lourd, complexe et infini, qui additionne en permanence les variations de courbure. C'est l'équivalent géométrique des nombres impairs de la fonction Zêta.

Face à cette lourdeur, le génie autodidacte Srinivasa Ramanujan a proposé une formule d'approximation prodigieuse. Sans utiliser l'infini, il a "tordu" $\pi$ et l'a marié aux demi-axes de l'ellipse ($a$ pour la longueur, $b$ pour la largeur) à l'intérieur d'une racine carrée :

$$P \approx \pi \left( 3(a+b) - \sqrt{(3a+b)(a+3b)} \right)$$

Dans le code qui suit, nous allons mesurer la précision de cette intuition fulgurante face à l'implacable vérité de l'intégrale exacte, sur une ellipse très étirée ($a=10$ et $b=2$).

```python
# @title
%%octave

a = 10;
b = 2;

P_rama = pi * ( 3*(a+b) - sqrt((3*a+b)*(a+3*b)) );

m = 1 - (b^2 / a^2);
[K, E] = ellipke(m);
P_exact = 4 * a * E;

erreur = abs(P_exact - P_rama);

disp('--- RAMANUJAN vs INTEGRALE ---')
disp('Valeur Ramanujan :')
disp(P_rama)
disp('Valeur Integrale exacte :')
disp(P_exact)
disp('Difference (Erreur) :')
disp(erreur)
```

# 3. La Synthèse : Le Cercle comme Cas Particulier

Notre perception humaine, née sur une sphère figée par la gravité, nous a fait croire que le cercle pur était la règle universelle, et l'ellipse une simple déformation.

La mécanique céleste et la cosmologie nous prouvent le contraire : l'ellipse est la véritable forme du mouvement libre dans l'univers. Le cercle parfait n'est qu'une anomalie statistique, un cas particulier où l'ellipse perd son élan.

Dans ce dernier calcul, nous allons observer cette "dégénérescence". Nous faisons grandir le petit axe ($b$) de notre ellipse jusqu'à ce qu'il rattrape le grand axe ($a$). Nous verrons alors la complexité transcendante de l'intégrale s'effondrer d'un coup pour nous rendre notre Invariant familier : $2\pi r$.

```python
# @title
%%octave

a = 10;

fprintf("=== LE RETOUR A L'INVARIANT (a = 10) ===\n\n");
fprintf("Nous elargissons l'ellipse pas a pas...\n\n");

% La boucle : b passe de 2 (tres aplati) a 10 (cercle parfait) par sauts de 2
for b = 2:2:10

    % Calcul Ramanujan
    P_rama = pi * ( 3*(a+b) - sqrt((3*a+b)*(a+3*b)) );

    % Calcul Exact
    m = 1 - (b^2 / a^2);
    [K, E] = ellipke(m);
    P_exact = 4 * a * E;

    % Affichage
    fprintf("Hauteur b = %2d | Formule exacte : %.4f | Ramanujan : %.4f\n", b, P_exact, P_rama);
end

fprintf("\nQuand b rattrape a (b=10), le ballet s'arrete.\n");
fprintf("La formule pure du cercle (2 * pi * r) reprend ses droits : %.4f\n", 2 * pi * a);
```

# 4. La Cellule Transitoire : Apéry et la rupture de la symétrie

Si le problème de Bâle ($\zeta(2)$) nous a enfermés dans l'illusion d'un univers circulaire régi par $\pi$, la constante d'Apéry ($\zeta(3)$) nous en libère violemment.

En 1978, face au scepticisme général, Roger Apéry a prouvé que la somme des inverses des cubes plonge dans une irrationalité totale. Dans cette dimension impaire, l'équité angulaire du cercle (et donc $\pi$) disparaît complètement. C'est l'équivalent arithmétique de l'ellipse qui refuse de se plier au rectangle.

Apéry n'a pas trouvé de "belle formule" géométrique. Il a dû forger de toutes pièces un outil algébrique d'une violence inouïe : des suites de nombres convergeant à une vitesse fulgurante pour prendre cette constante en tenaille et prouver qu'elle ne s'arrêterait jamais.

Observons cette chute dans l'irrationalité, là où le cercle ne peut plus rien pour nous.

```python
# @title
%%octave

fprintf("=== LA CELLULE TRANSITOIRE : APERY ET ZETA(3) ===\n\n");

% 1. La limite du calcul humain (force brute)
N = 1000000;
n = 1:N;
somme_impaire = sum(1 ./ (n.^3));

% 2. La valeur reconnue de la constante d'Apery
apery_const = 1.202056903159594;

fprintf("L'effondrement de la symetrie circulaire...\n");
fprintf("Aucun lien avec pi n'est possible ici.\n\n");

fprintf("Somme calculee (force brute) : %.9f\n", somme_impaire);
fprintf("Constante d'Apery (theorique): %.9f\n", apery_const);

erreur = abs(apery_const - somme_impaire);
fprintf("Ecart vers l'infini          : %.9f\n\n", erreur);

fprintf("Le cercle est vaincu. L'univers complexe s'ouvre.\n");
```

```python
# @title
%%octave

fprintf("=== LE DUEL FINAL : FORCE BRUTE vs LE GENIE D'APERY ===\n\n");

% Seulement 15 petites etapes de calcul !
N = 15;

% 1. La force brute (L'addition lente classique)
n_brute = 1:N;
zeta3_brute = sum(1 ./ (n_brute.^3));

% 2. Le piege d'Apery (L'accelerateur algebrique)
zeta3_apery = 0;
for k = 1:N
    numerateur = (-1)^(k-1);
    denominateur = (k^3) * nchoosek(2*k, k); % nchoosek est le coefficient binomial
    zeta3_apery = zeta3_apery + (numerateur / denominateur);
end
zeta3_apery = zeta3_apery * (5/2);

% 3. La vraie valeur transcendante (pour l'arbitrage)
vraie_valeur = 1.202056903159594;

% --- L'HEURE DU VERDICT ---
fprintf("Apres avoir transpire sur %d etapes de calcul :\n\n", N);

fprintf("1. La Somme Classique (Le carre humain) : %.9f\n", zeta3_brute);
fprintf("   Il lui manque encore                 : %.9f pour atteindre la verite.\n\n", abs(vraie_valeur - zeta3_brute));

fprintf("2. Le Piege d'Apery (La tenaille)       : %.9f\n", zeta3_apery);
fprintf("   L'erreur n'est plus que de           : %.9f !!!\n\n", abs(vraie_valeur - zeta3_apery));

fprintf("La force brute s'epuise. Le genie algebrique a deja capture l'infini.\n");
```
